```python
# Understanding missingness in each column. 

missing_percentage = (df.isna().sum() / len(df)) * 100

columns_with_missing = missing_percentage[missing_percentage > 0]

print("Columns with missing values (percentage):")
print(columns_with_missing)
```

### There are 4 columns with missing values.
- The agent column has 16340 (13.69%) missing values.
- The company column has 112593 (94.31%)missing values.
- The country column has 488 (0.41%)missing values.
- The children column has 4 (0.0034%) missing values.



```python
# Unique Values in each column 
def unique_columns(df):
    for col in df.columns:
        uniques = df[col].unique()
        print(f"Number of Unique values: {len(uniques)}")
        print(f"Unique Values: {col}\n:{uniques}\n")

unique_columns(df)
```
